```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
```

```python
fountains = gpd.read_file('../data/Public Water Fountains and Hydration Systems (GIS).geojson')
print(fountains.crs)
fountains.head( )
```

```python
zipcodes = gpd.read_file('../data/zipcodes.geojson')
print(zipcodes.crs)
zipcodes.head( )
```

```python
zipcodes = zipcodes[['zip', 'po_name', 'geometry']]
zipcodes.head()
```

```python
fountains_geo = gpd.GeoDataFrame(fountains, 
                           crs = zipcodes.crs, 
                           geometry = gpd.points_from_xy(fountains.lon, fountains.lat))
```

```python
fountains_geo.head()
```

```python
fountains_by_zip = gpd.sjoin(fountains_geo, zipcodes, predicate = 'within')
```
